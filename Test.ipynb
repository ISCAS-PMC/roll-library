```groovy
import roll.jupyter.*;
import roll.words.Alphabet;
import java.util.ArrayList;
import roll.automata.NBA;

ArrayList<Character> apList = new ArrayList<>();
apList.add((char)'a');
apList.add((char)'b');

JupyterROLL.createAlphabet(apList);

nba = JupyterROLL.createNBA();
nba.createState();
nba.createState();
nba.createState();
Alphabet alphabet = nba.getAlphabet();
// add transitions for NBA recognizing a^w + ab^w
int fst = 0, snd = 1, thd = 2;
nba.getState(fst).addTransition(alphabet.indexOf((char)'a'), snd);
nba.getState(fst).addTransition(alphabet.indexOf((char)'a'), thd);
nba.getState(snd).addTransition(alphabet.indexOf((char)'a'), snd);
nba.getState(thd).addTransition(alphabet.indexOf((char)'b'), thd);
nba.setInitial(fst);
nba.setFinal(snd);
nba.setFinal(thd);

seq = JupyterROLL.learningSeq("recurrent", "table", nba);
seq.get(1)
```

```groovy
import roll.jupyter.*;
import java.util.function.BiFunction;
import roll.words.*;

// learn (a|b)^* b^w
mqOracle = {stem,loop -> 
    if (loop.length() < 1) return false;
    for (int i = 0; i < loop.length(); i++) {
        if (loop.charAt(i) != 'b') {
                return false;
        }
    }
    return true;
};
    
nbaLearner = JupyterROLL.createNBALearner("periodic", "table", mqOracle);
nbaLearner.getHypothesis();
```

```groovy
nbaLearner.refineHypothesis("", "b");
nbaLearner.getHypothesis();
```

```groovy
nbaLearner.refineHypothesis("", "ba");
nbaLearner.getHypothesis();
```

```groovy
import roll.jupyter.*;
import java.util.function.Function;
import roll.words.*;

mqOracle = {loop -> 
    // 4n + 3 -> b
    int num = 0;
    for (int i = 0; i < loop.length(); i++) {
        if (loop.charAt(i) == 'b') {
                num ++;
        }
    }
    if(num % 4 == 3) return true;
    return false;
};
    
dfaLearner = JupyterROLL.createDFALearner("column", "tree", mqOracle);
dfaLearner.getHypothesis()
```

```groovy
dfaLearner.refineHypothesis("bbb");
dfaLearner.getHypothesis()
```
